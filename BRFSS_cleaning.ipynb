```python
cols_to_keep = ['SEQNO', 'MENTHLTH', 'DEAF', 'BLIND', 'DECIDE', 'DIFFWALK', 'DIFFDRES', 'DIFFALON', 'DIABETE4', 'GENHLTH', 
                'PHYSHLTH', 'BPHIGH6', 'CVDINFR4', 'CVDCRHD4', 'CVDSTRK3', 'ASTHMA3', 'CHCSCNCR', 'CHCOCNCR', 
                'CHCCOPD3', 'CHCKDNY2', 'HAVARTH5', 'MARITAL', 'EDUCA', 'RENTHOM1', 'VETERAN3', 'EMPLOY1', 'CHILDREN', 'INCOME3', 
                '_BMI5', 'SEXVAR', '_IMPRACE', '_AGEG5YR', '_RFSMOK3', '_CURECI1', '_DRNKWK1', 'POTADA1_', 'FTJUDA2_', 'FRUTDA2_',
                'GRENDA1_', 'VEGEDA2_', 'MEDCOST1', 'PRIMINSR', 'EXERANY2']
len(cols_to_keep)
```

```python
df = BRFSS[cols_to_keep].copy()
```

```python
pd.set_option('display.max_columns', 52)
df
```

```python
for i in df.columns:
    print(df[i].value_counts())
```

```python
#cleaning mental health days
df['MENTHLTH'][df['MENTHLTH'] == 88] = 0
df
```

```python
df.drop(df[df['MENTHLTH'] > 31].index, inplace = True)
df
```

```python
disability = ['DEAF', 'BLIND', 'DECIDE', 'DIFFWALK', 'DIFFDRES', 'DIFFALON']

for i in disability:
    df[i][df[i] == 2] = 0
    df.drop(df[df[i] > 2].index, inplace = True)

df
```

```python
pd.set_option('display.max_columns', 53)
df['DISCORE'] = df[list(df[disability])].sum(axis=1)
df
```

```python
#test_dataset = df.copy()
bad_health = ['DIABETE4', 'GENHLTH', 'BPHIGH6', 'CVDINFR4', 'CVDCRHD4', 'CVDSTRK3', 'ASTHMA3', 'CHCSCNCR', 
              'CHCOCNCR', 'CHCCOPD3', 'CHCKDNY2', 'HAVARTH5']
bad_health_special = ['PHYSHLTH']
#for i in bad_health:
 #   print(df[i].value_counts())
    
for i in bad_health:
    df[i][df[i] == 2] = 0
    df.drop(df[df[i] >= 7].index, inplace = True)
    print(df.shape)

df

```

```python
df = df.rename(columns = {'GENHLTH': 'HEALTHSCORE', 
                          'BPHIGH6': 'HIGHBP', 
                          'CVDINFR4': 'HEARTATTACK', 
                          'CVDCRHD4': 'HEARTDISEASE', 
                          'CVDSTRK3': 'STROKE', 
                          'ASTHMA3': 'ASTHMA', 
                          'CHCSCNCR': 'SKINCANCER', 
                          'CHCOCNCR': 'OTHERCANCER', 
                          'CHCCOPD3': 'LUNGISSUE', 
                          'CHCKDNY2': 'KIDNEYISSUE', 
                          'HAVARTH5': 'ATHRITIS'})
```

```python
df['PHYSHLTH'][df['PHYSHLTH'] == 88] = 0
df.drop(df[df['PHYSHLTH'] >= 31].index, inplace = True)
df
```

```python
df['DIABETE4'][df['DIABETE4'] == 3] = 0
df['DIABETE4'][df['DIABETE4'] == 4] = 0
df = df.rename(columns = {'DIABETE4': 'DIAB'})
df
```

```python
demos = ['MARITAL', 'EDUCA', 'RENTHOM1', 'VETERAN3', 'EMPLOY1', 'CHILDREN', 'INCOME3', 
         '_BMI5', 'SEXVAR', '_IMPRACE', '_AGEG5YR']

for i in demos:
    print(df[i].value_counts())
```

```python
for i in range(2, 7):
    if i == 6:
        df['MARITAL'][df['MARITAL'] == i] = 1
    else:
        df['MARITAL'][df['MARITAL'] == i] = 0
df.drop(df[df['MARITAL'] == 9].index, inplace = True)
df = df.rename(columns = {'MARITAL': 'PARTNERED'})

df['PARTNERED'].value_counts()
```

```python
df['EDUCA'][df['EDUCA'] <= 5] = 0
df['EDUCA'][df['EDUCA'] == 6] = 1
df.drop(df[df['EDUCA'] == 9].index, inplace = True)
df = df.rename(columns = {'EDUCA': 'COLLEGE'})
df['COLLEGE'].value_counts()
```

```python
df['RENTHOM1'][df['RENTHOM1'] == 2] = 0
df['RENTHOM1'][df['RENTHOM1'] == 3] = 0
df.drop(df[df['RENTHOM1'] == 7].index, inplace = True)
df.drop(df[df['RENTHOM1'] == 9].index, inplace = True)
df = df.rename(columns = {'RENTHOM1': 'OWNHOME'})
df['OWNHOME'].value_counts()
```

```python
df.shape
```

```python
df['VETERAN3'][df['VETERAN3'] == 2] = 0
df.drop(df[df['VETERAN3'] == 7].index, inplace = True)
df.drop(df[df['VETERAN3'] == 9].index, inplace = True)
df = df.rename(columns = {'VETERAN3': 'VET'})
df['VET'].value_counts()
```

```python
df['EMPLOY1'][df['EMPLOY1'] == 2] = 1
df['EMPLOY1'][df['EMPLOY1'] == 7] = 2
df['EMPLOY1'][df['EMPLOY1'] == 3] = 0
df['EMPLOY1'][df['EMPLOY1'] == 4] = 0
df['EMPLOY1'][df['EMPLOY1'] == 5] = 0
df['EMPLOY1'][df['EMPLOY1'] == 6] = 0
df['EMPLOY1'][df['EMPLOY1'] == 8] = 3
df.drop(df[df['EMPLOY1'] == 7].index, inplace = True)
df.drop(df[df['EMPLOY1'] == 9].index, inplace = True)
df = df.rename(columns = {'EMPLOY1': 'EMPLOYED'})
df['EMPLOYED'].value_counts()
```

```python
df = pd.concat([df, pd.get_dummies(df['EMPLOYED']).rename(columns = {0: 'UNEMP', 1: 'EMPL', 2:'RETIR', 3:'UNAB_WORK'})], axis = 1)
```

```python
df['CHILDREN'][df['CHILDREN'] == 88] = 0
df.drop(df[df['CHILDREN'] == 99].index, inplace = True)
df.drop(df[df['CHILDREN'] > 15].index, inplace = True)
df['CHILDREN'][df['CHILDREN'] > 0] = 1
df['CHILDREN'].value_counts()
```

```python
df.drop(df[df['INCOME3'] == 77].index, inplace = True)
df.drop(df[df['INCOME3'] ==99].index, inplace = True)
df = df.rename(columns = {'INCOME3': 'INCOMECAT'})
df['INCOMECAT'].value_counts()
```

```python
df['_BMI5'].value_counts()
df = df.rename(columns = {'_BMI5': 'BMI'})
```

```python
df = pd.concat([df, pd.get_dummies(df['SEXVAR']).rename(columns = {1: 'M', 2: 'F'})], axis = 1)
df.drop(['SEXVAR'], axis=1, inplace = True)
df
```

```python
df = pd.concat([df, pd.get_dummies(df['_IMPRACE']).rename(columns = {1: 'WHITE', 2: 'BLACK', 3: 'ASIAN', 4: 'AM_IND', 5: 'HISP', 6: 'ANOTHER_RACE'})], axis = 1)
df.drop(['_IMPRACE'], axis=1, inplace = True)
df
```

```python
df.drop(df[df['_AGEG5YR'] ==14].index, inplace = True)
df['_AGEG5YR'].value_counts()
df = df.rename(columns = {'_AGEG5YR': 'AGEG5YR'})
```

```python
risk_behavior = ['_RFSMOK3', '_CURECI1', '_DRNKWK1', 'POTADA1_', 'MEDCOST1']

for i in risk_behavior:
    print(df[i].value_counts())
```

```python
df['_RFSMOK3'][df['_RFSMOK3'] == 1] = 0
df['_RFSMOK3'][df['_RFSMOK3'] == 2] = 1
df.drop(df[df['_RFSMOK3'] == 9].index, inplace = True)
df = df.rename(columns = {'_RFSMOK3': 'SMOKE'})
df['SMOKE'].value_counts()
```

```python
df['_CURECI1'][df['_CURECI1'] == 1] = 0
df['_CURECI1'][df['_CURECI1'] == 2] = 1
df.drop(df[df['_CURECI1'] == 9].index, inplace = True)
df = df.rename(columns = {'_CURECI1': 'ECIG'})
df['ECIG'].value_counts()
```

```python
df.drop(df[df['_DRNKWK1'] == 99900].index, inplace = True)
df = df.rename(columns = {'_DRNKWK1': 'DRINKPERW'})
df['DRINKPERW'].value_counts()
```

```python
df = df.rename(columns = {'POTADA1_': 'POTADAY'})
nan_count = df['POTADAY'].isna().sum()
print(nan_count)
```

```python
df['MEDCOST1'][df['MEDCOST1'] == 2] = 0
df.drop(df[df['MEDCOST1'] == 7].index, inplace = True)
df.drop(df[df['MEDCOST1'] == 9].index, inplace = True)
df = df.rename(columns = {'MEDCOST1': 'AFFORDOC'})
df['AFFORDOC'].value_counts()
```

```python
nan_count = df.isna().sum()
print(nan_count)
```

```python
buffers = ['FTJUDA2_', 'FRUTDA2_', 'GRENDA1_', 'VEGEDA2_', 'PRIMINSR', 'EXERANY2']
for i in buffers:
    print(df[i].value_counts())
```

```python
df = df.rename(columns = {'FTJUDA2_': 'FRUIT_J_DAY'})
df = df.rename(columns = {'FRUTDA2_': 'FRUITDAY'})
df = df.rename(columns = {'GRENDA1_': 'GRNVEGDAY'})
df = df.rename(columns = {'VEGEDA2_': 'OTHERVEGDAY'})
```

```python
for i in range(1,11):
    df['PRIMINSR'][df['PRIMINSR'] == i] = 1
df['PRIMINSR'][df['PRIMINSR'] == 88] = 0
df['PRIMINSR'][df['PRIMINSR'] == 2] = 0
df.drop(df[df['PRIMINSR'] == 77].index, inplace = True)
df.drop(df[df['PRIMINSR'] == 99].index, inplace = True)
df = df.rename(columns = {'PRIMINSR': 'INSURED'})
df['INSURED'].value_counts()
```

```python
df['EXERANY2'][df['EXERANY2'] == 2] = 0
df.drop(df[df['EXERANY2'] == 7].index, inplace = True)
df.drop(df[df['EXERANY2'] == 9].index, inplace = True)
df = df.rename(columns = {'EXERANY2': 'EXERCISE'})
df['EXERCISE'].value_counts()
```
