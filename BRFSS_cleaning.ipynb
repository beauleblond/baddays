```python
for i in df.columns:
    print(df[i].value_counts())
```

```python
#cleaning mental health days
df['MENTHLTH'][df['MENTHLTH'] == 88] = 0
df
```

```python
df.drop(df[df['MENTHLTH'] > 31].index, inplace = True)
df
```

```python
disability = ['DEAF', 'BLIND', 'DECIDE', 'DIFFWALK', 'DIFFDRES', 'DIFFALON']

for i in disability:
    df[i][df[i] == 2] = 0
    df.drop(df[df[i] > 2].index, inplace = True)

df
```

```python
pd.set_option('display.max_columns', 53)
df['DISCORE'] = df[list(df[disability])].sum(axis=1)
df
```

```python
#test_dataset = df.copy()
bad_health = ['DIABETE4', 'GENHLTH', 'BPHIGH6', 'CVDINFR4', 'CVDCRHD4', 'CVDSTRK3', 'ASTHMA3', 'CHCSCNCR', 
              'CHCOCNCR', 'CHCCOPD3', 'CHCKDNY2', 'HAVARTH5']
bad_health_special = ['PHYSHLTH']
#for i in bad_health:
 #   print(df[i].value_counts())
    
for i in bad_health:
    df[i][df[i] == 2] = 0
    df.drop(df[df[i] >= 7].index, inplace = True)
    print(df.shape)

df

```

```python
df['PHYSHLTH'][df['PHYSHLTH'] == 88] = 0
df.drop(df[df['PHYSHLTH'] >= 31].index, inplace = True)
df
```

```python
df['DIABETE4'][df['DIABETE4'] == 3] = 0
df['DIABETE4'][df['DIABETE4'] == 2] = 0
df['DIABETE4'][df['DIABETE4'] == 4] = 0
df = df.rename(columns = {'DIABETE4': 'DIAB'})
df
```

```python
demos = ['MARITAL', 'EDUCA', 'RENTHOM1', 'VETERAN3', 'EMPLOY1', 'CHILDREN', 'INCOME3', 
         '_BMI5', 'SEXVAR', '_IMPRACE', '_AGEG5YR']

for i in demos:
    print(df[i].value_counts())
```

```python
for i in range(2, 7):
    df['MARITAL'][df['MARITAL'] == i] = 0
df.drop(df[df['MARITAL'] == 9].index, inplace = True)
df = df.rename(columns = {'MARITAL': 'MARRIED'})

df['MARRIED'].value_counts()
```

```python
df['EDUCA'][df['EDUCA'] <= 5] = 0
df['EDUCA'][df['EDUCA'] == 6] = 1
df.drop(df[df['EDUCA'] == 9].index, inplace = True)
df = df.rename(columns = {'EDUCA': 'COLLEGE'})
df['COLLEGE'].value_counts()
```

```python
df['RENTHOM1'][df['RENTHOM1'] == 2] = 0
df['RENTHOM1'][df['RENTHOM1'] == 3] = 0
df.drop(df[df['RENTHOM1'] == 7].index, inplace = True)
df.drop(df[df['RENTHOM1'] == 9].index, inplace = True)
df = df.rename(columns = {'RENTHOM1': 'OWNHOME'})
df['OWNHOME'].value_counts()
```

```python
df.shape
```

```python
df['VETERAN3'][df['VETERAN3'] == 2] = 0
df.drop(df[df['VETERAN3'] == 7].index, inplace = True)
df.drop(df[df['VETERAN3'] == 9].index, inplace = True)
df = df.rename(columns = {'VETERAN3': 'VET'})
df['VET'].value_counts()
```

```python
df['EMPLOY1'][df['EMPLOY1'] == 2] = 1
df['EMPLOY1'][df['EMPLOY1'] == 7] = 2
df['EMPLOY1'][df['EMPLOY1'] == 3] = 0
df['EMPLOY1'][df['EMPLOY1'] == 4] = 0
df['EMPLOY1'][df['EMPLOY1'] == 5] = 0
df['EMPLOY1'][df['EMPLOY1'] == 6] = 0
df['EMPLOY1'][df['EMPLOY1'] == 8] = 3
df.drop(df[df['EMPLOY1'] == 7].index, inplace = True)
df.drop(df[df['EMPLOY1'] == 9].index, inplace = True)
df = df.rename(columns = {'EMPLOY1': 'EMPLOYED'})
df['EMPLOYED'].value_counts()
```

```python
df = pd.concat([df, pd.get_dummies(df['EMPLOYED']).rename(columns = {0: 'UNEMP', 1: 'EMPL', 2:'RETIR', 3:'UNAB_WORK'})], axis = 1)
```

```python
df['CHILDREN'][df['CHILDREN'] == 88] = 0
df.drop(df[df['CHILDREN'] == 99].index, inplace = True)
df.drop(df[df['CHILDREN'] > 15].index, inplace = True)
df['CHILDREN'][df['CHILDREN'] > 0] = 1
df['CHILDREN'].value_counts()
```

```python
df.drop(df[df['INCOME3'] == 77].index, inplace = True)
df.drop(df[df['INCOME3'] ==99].index, inplace = True)
df['INCOME3'].value_counts()
```
